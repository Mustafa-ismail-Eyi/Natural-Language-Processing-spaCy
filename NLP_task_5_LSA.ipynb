## 2. Soru 5 kereden fazla geçen Bigram'lar ve Trigram'ları bulma

```python
#aksi takdirde hata veriyor

corpus=" ".join([str(x[-1] +"\n") for x in [sentence.split('\t') for sentence in tr_corpus.split('\n')]])
#corpus= corpus.replace(',','')
#corpus=corpus.lower()
nlp_tr.max_length = len(corpus)
doc = nlp_tr(corpus)
```

```python
def tokenize(doc):
    tokens_of_word = list()
    for tokens in doc:
        if not(tokens.is_stop or tokens.is_punct or tokens.is_space or tokens.is_right_punct or tokens.like_num or tokens.shape=='X' or tokens.shape=='x' or tokens.is_quote):
            tokens_of_word.append(tokens)
    
    #print(tokens_of_word)
    return tokens_of_word
```

```python
# spaCy kendi içerisinde bir ngram, bigram veya trigram fonksiyonu içermediği için biz oluşturduk.
def ngram(document, n):
    """document: list, array like object,
    n: n of ngram"""
    ngrams_  = list()
    for i in range(len(document)-n+1):
        ngrams_.append(document[i:i+n])
    return ngrams_
```

```python
# Oluturulan ngram listelerinin içindeki unique elemanları sayan fonksiyon
def ngram_counter(ngram_list,threshold=0):
    """ngram_list: ngram list of that you created
       threshold: the count of ngrams that you want"""
    listed = list()
    listed=([ngrams,ngram_list.count(ngrams)] for ngrams in ngram_list if ngram_list.count(ngrams)>=threshold)
    unique = list()
    for x in listed:
        if x not in unique:
            unique.append(x)
    return unique
```

```python
tokenized_doc = tokenize(doc)
```

```python
bigrams = ngram(tokenized_doc, 2)
trigrams = ngram(tokenized_doc, 3)
```

```python
counts_birgrams = ngram_counter(bigrams, 5)
```

```python
sorted(counts_birgrams, key=lambda counts_birgrams:counts_birgrams[1],reverse=True)[0:10]
```

```python
counts_trigrams = ngram_counter(trigrams, 5)
```

## 3. Soru Latent Semantic Analysis

```python
lemma_of_sentence = list()
for sentences in corpus.split('\n'):
    doc = nlp_tr(sentences)
    lemmas = list()
    for x in doc:
        if not(x.is_punct or x.like_num or not(x.has_vector) or x.is_space):
            lemmas.append(x.lemma_)
    lemma_of_sentence.append((sentences,str(" ".join(lemmas)).lower()))
```

```python
lemma_of_sentence
```

```python
df = pd.DataFrame(data=lemma_of_sentence,columns=['sentences','lemma'])
df.tail()
```

```python
df.drop([10000,10001],axis=0,inplace=True)
df.tail()
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,3))
X = vectorizer.fit_transform(df.lemma)
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=10)
LSA_X=svd.fit_transform(X)
```

```python
def get_topics(components, feature_names, n=2):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx))
        print([(feature_names[i], topic[i])
                        for i in topic.argsort()[:-n - 1:-1]])
get_topics(svd.components_, vectorizer.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim=cosine_similarity(LSA_X,LSA_X)
```

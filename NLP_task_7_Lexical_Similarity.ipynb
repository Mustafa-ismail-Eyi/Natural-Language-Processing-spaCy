```python
review = " ".join(data[data['URL']=='https://www.hepsiburada.com/logitech-m175-kablosuz-nano-mouse-910-002777-p-BD40120']['Review'][:800].values)
```

```python
nlp_tr = spacy.load('models/model-best')
```

```python
doc = nlp_tr(review)
```

```python
# stop words'lerden arındırma
def filter_doc():
    filtered_list_doc = list()
    global doc
    for tokens in doc:
        if not(tokens.text[0].isdigit() or tokens.text[-1].isdigit()):
            if not(tokens.is_punct or tokens.is_stop or tokens.is_space):
                filtered_list_doc.append(tokens.text.lower())
    return filtered_list_doc
doc_list=filter_doc()
```

```python
# kelime ve sayılarını alma
count_of_filtered_doc = dict(Counter(str(token) for token in doc_list))
```

```python
# dataframe'e dönüştürme
words_df=pd.DataFrame(list(zip(count_of_filtered_doc.keys(),count_of_filtered_doc.values())),columns=['words','counts'])
```

```python
# en sık geçen yani 5 kereden fazla geçen kelimeler tespit edildi
# görev kapsamında bu keliemeler yazım bakımından doğru olan keliemeler olarak kabul edildi
most_freq = words_df[words_df['counts']>=5].sort_values(by='counts',ascending=False)
```

```python
# daha seyrek geçen kelimeler tespit edildi yani 5 kereden az geçen keliemelr tespit edildi
# görev kapsamında bu kelimeler yazım hatası yapılmış keliemeler olarak kabul edildi
least_freq = words_df[words_df['counts']<5].sort_values(by='counts',ascending=False)
```

```python
# Seyrek geçen herbir kelimenin sık geçen her bir kelimeye göre en benzeri Jaccard similarity'e göre belirlendi 
# jaccar similarity elde etmek için -> (1 - Jaccard_Distance)
column_names = ["error", "true", "similarity_ratio"]
most_similar_ones = pd.DataFrame(columns = column_names)


def get_most_similars(dataframe):
    global most_similar_ones
    most_similar_ones= most_similar_ones.append(dataframe)

def jaccard_similarities():
    global column_names

    for y in least_freq['words'].values:
        error_word = []
        true_word = []
        similarities = []
        
        temporary_df = pd.DataFrame(columns = column_names)

        for x in most_freq['words'].values:
            #print('{first} and {second} similarities'.format(first=y,second=x),1-jaccard(y, x))
            error_word.append(y)
            true_word.append(x)
            similarities.append(1-jaccard(y, x))
        temporary_df['error'] = error_word
        temporary_df['true'] = true_word
        temporary_df['similarity_ratio'] = similarities
        
        get_most_similars(temporary_df.sort_values(by='similarity_ratio',ascending=False)[0:1])
        

jaccard_similarities()
```

```python
# elde ettiğimiz sonuçların ilk 50 örneği aşağıda görüldüğü gibidir
most_similar_ones.sort_values(by='similarity_ratio',ascending=False)[:50]
```

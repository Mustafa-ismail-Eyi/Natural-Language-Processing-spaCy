```python
import spacy
```

```python
nlp = spacy.load('models\model-best')
```

```python
doc = nlp("Kırmızı kapaklı kitap.")
```

```python
print(f"TOKENS\t\tTAG")
for tokens in doc:
    print(f"{tokens}\t\t{tokens.tag_}")
```

```python
f = open('tur-tr_web_2019_10K/tur-tr_web_2019_10K-sentences.txt','r',encoding='utf8')
```

```python
tr_corpus=f.read()
f.close()
```

```python
sentence_list = tr_corpus.split('\t')[0:100]
corpus = " ".join(sentence_list)
```

```python
print(corpus)
```

```python
doc = nlp(corpus)
```
